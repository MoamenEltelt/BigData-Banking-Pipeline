```python
import pandas as pd

df = pd.read_csv('../data/Banking_Transactions_USA_2023_2024.csv')

df.head()
```

```python
df.info()
```

```python
df['Transaction_Type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Transaction_Type'].value_counts().plot(kind='bar')

plt.title("Transaction Type Count")
plt.xlabel("Transaction Type")
plt.ylabel("Count")

plt.show()
```

```python
df['Category'].value_counts()
```

```python
df['Category'].value_counts().plot(kind='bar', figsize=(12,5))

plt.title("Transaction Categories")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()
```

```python
df['City'].value_counts()
```

```python
df['City'].value_counts().plot(kind='bar', figsize=(12,5))

plt.title("Transactions by City")
plt.xlabel("City")
plt.ylabel("Count")

plt.show()
```

```python
df['Fraud_Flag'].value_counts()
```

```python
df['Fraud_Flag'].value_counts().plot(kind='bar')

plt.title("Fraud Transactions")
plt.xlabel("Fraud")
plt.ylabel("Count")

plt.show()
```
